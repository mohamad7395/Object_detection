```python
import matplotlib.pyplot as plt
from pandas.core.common import flatten
import copy
import numpy as np
import random

import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
from torchvision import datasets, transforms, models
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2

import glob
from tqdm import tqdm
import pickle

f = open('/Users/mohamad/python-workspace/lab/extract_img/labels.plk', 'rb')
labels_p = pickle.load(f)   
transform = T.Compose([
                       T.ToTensor(),
                      T.Resize(36)])

####################################################
#       Create Train, Valid and Test sets
####################################################
train_data_path = '/Users/mohamad/python-workspace/lab/extract_img/tr_pic2/train' 
test_data_path = '/Users/mohamad/python-workspace/lab/extract_img/tr_pic2/test' 

train_folder_paths = [] #to store image paths in list

#1.
# get all the paths from train_data_path and append image paths and class to to respective lists
# eg. train path-> 'images/train/26.Pont_du_Gard/4321ee6695c23c7b.jpg'
# eg. class -> 26.Pont_du_Gard
for data_path in glob.glob(train_data_path + '/*'):
    train_folder_paths.append(data_path)
    

# print('train_image_path example: ', train_folder_paths)


#2.
# split train valid from train paths (80,20)
# train_image_paths, valid_image_paths = train_folder_paths[:int(0.8*len(train_folder_paths))], train_folder_paths[int(0.8*len(train_folder_paths)):] 

#3.
# create the test_image_paths
test_folder_paths = []
for data_path in glob.glob(test_data_path + '/*'):
    test_folder_paths.append(data_path)

test_folder_paths = sorted(test_folder_paths)

train_folder_paths = sorted(train_folder_paths)

# print("Train size: {}\nValid size: {}\nTest size: {}".format(len(train_folder_paths), len(valid_image_paths), len(test_image_paths)))
  
    

class imgDataset(Dataset):
    def __init__(self, image_paths,seq_len, transform=False):
        self.image_paths = image_paths
        self.transform = transform
        self.seq_len = seq_len
        
    def __len__(self):
        return len(self.image_paths)-self.seq_len

    def __getitem__(self, idx):
        start = idx
        end = idx + self.seq_len
        images = []
        labels = []
        indices = list(range(start,end))
        folder_filepath=self.image_paths[idx]
        image_paths=sorted(glob.glob(folder_filepath + '/*'))
        
        for i in indices:
            
            image_filepath = self.image_paths[i]
            
            image = cv2.imread(image_filepath)
            image = self.transform(image)
            images.append(image)
#         image = cv2.resize(image, (90,160), interpolation = cv2.INTER_AREA)
#         image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#         image = cv2
            indx=int(image_filepath.split('/')[-1].split('.')[-2])
            row=int(image_filepath.split('/')[-2])   
            label=(labels_p[row][indx])/10
            labels.append(torch.tensor(label))
            

        # label = image_filepath.split('/')[-2]
        # label = class_to_idx[label]

        images = torch.stack(images)
        labels = torch.stack(labels)
        
        return images, labels 
    
#######################################################
#                  Create Dataset
#######################################################
seq_l=5
train_dataset = imgDataset(train_image_paths,seq_l, transform)
valid_dataset = imgDataset(valid_image_paths,seq_l, transform) #test transforms are applied
test_dataset = imgDataset(test_image_paths,seq_l, transform)


# print(test_dataset[0])

#######################################################
#                  Define Dataloaders
#######################################################
batch_size = 4
train_dl = DataLoader(
    train_dataset, batch_size=batch_size, shuffle=False
)

valid_dl = DataLoader(
    valid_dataset, batch_size=batch_size, shuffle=False
)


test_dl = DataLoader(
    test_dataset, batch_size=batch_size, shuffle=False
)

print(batch_size)

```

```python
# Load in relevant libraries, and alias where appropriate
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms

# Define relevant variables for the ML task

num_classes = 64
learning_rate = 0.001
num_epochs = 20

# Device will determine whether to run the training on GPU or CPU.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
#Defining the convolutional neural network
class LeNet5(nn.Module):
    def __init__(self, num_classes):
        super(LeNet5, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(3, 6, kernel_size=5, stride=1, padding=0),
            nn.BatchNorm2d(6),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(6, 12, kernel_size=5, stride=1, padding=0),
            nn.BatchNorm2d(12),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.fc = nn.Linear(936, 256)
        self.relu = nn.ReLU()
        self.fc1 = nn.Linear(256, 128)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(128, 64)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        out = self.relu(out)
        out = self.fc1(out)
        out = self.relu1(out)
        out = self.fc2(out)
        return out
```

```python
model = LeNet5(num_classes).to(device)

#Setting the loss function

cost = nn.MSELoss()
#Setting the optimizer with the model parameters and learning rate
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

#this is defined to print how many steps are remaining when training
total_step = len(train_dl)

```

```python
def yhat(seq_l,f):
    aa = 3750
    d_t = 0.1
    one = np.ones(seq_l)
    a_t = np.arange(1,seq_l+1)*d_t
    A = np.vstack((a_t,one)).T
    # np.dot(A.T,A)
    a = a_t**2*aa
#     y_h = a + np.dot(np.dot(A,(np.linalg.inv(np.dot(A.T,A)))),A.T)*(f-a)
    y_hh = a + np.dot(.dot(np.dot(A,(np.linalg.inv(np.dot(A.T,A)))),A.T),(f-a))
    return y_hh
```

```python
requires_grad=True
total_step = len(train_dl)
t_loss = []
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_dl):  
        images = images.to(device)
        images = torch.reshape(images,(images.shape[0]*seq_l,3,64,36))
        labels = labels.to(device).to(torch.long)
#         labels = torch.flatten(labels)
        #Forward pass
        
        outputs = model(images)
#         print(outputs)
        
        _, out = torch.max(torch.absolute(outputs.data), 1)
        out=torch.reshape(out,(int(images.shape[0]/seq_l),seq_l))
#         print(out)
#         print(out.size())
#         print(out.size())
        loss_v =[]
        for i in range(int(images.shape[0]/seq_l)):
# #           
            yh=yhat(seq_l,(out[i]).detach().numpy())
#             print(yh)
            loss_v.append(cost(torch.tensor(yh,requires_grad=True), out[i].to(torch.double)))
            
#             loss_v.append(cost(out[i],labels[i]))
        loss=sum(loss_v)/batch_size
#         print(loss)
#         print(float(loss_v[0]))
        	
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        		
#         if (i+1) % 400 == 0:
#             print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
#         		           .format(epoch+1, num_epochs, i+1, total_step, loss.item()))
    print("loss",epoch,"=====",float(loss))    
    t_loss.append(float(loss))
from matplotlib.pyplot import figure
    
figure(figsize=(20, 11), dpi=220)    
plt.plot(t_loss)    
```

```python
# Test the model
# In test phase, we don't need to compute gradients (for memory efficiency)
pred = []
pred = torch.tensor(pred)
targets = []
targets = torch.tensor(targets)
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in test_dl:
        images = images.to(device)
        images = torch.reshape(images,(images.shape[0]*seq_l,3,64,36))
        labels = labels.to(device).to(torch.long)
        labels = torch.flatten(labels)
        print(labels)
        targets = torch.cat((targets,labels),0)
#         print(labels)
        
        outputs = model(images)
        print(outputs.size())
#         print(outputs)
        _, predicted = torch.max(outputs.data, 1)
        
        pred = torch.cat((pred,predicted),0)
        print(pred.size())
#         print(predicted)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print('Accuracy of the network on the 10000 test images: {} %'.format(100 * correct / total))
	 
from matplotlib.pyplot import figure

figure(figsize=(20, 11))
# figure.savefig('test2png.png', dpi=200)
plt.plot(targets,marker='o')
plt.plot(pred,marker='o')
plt.xlabel("Test Set")
plt.ylabel("Height")
plt.legend(["True Label","Prediction"])
plt.savefig("result_cl")
print(pred-targets)
print(sum(np.abs(pred-targets)))
print(cost(targets,pred))
from sklearn.metrics import mean_absolute_error
MAE_svr = mean_absolute_error(targets,pred)
err_percent = (MAE_svr/targets.mean())*100
print('error = ',100-err_percent,'%' )
```
